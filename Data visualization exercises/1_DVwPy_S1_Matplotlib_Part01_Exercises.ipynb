```python
weather = pd.read_csv("weatherAUS.csv")
pd.options.display.max_columns = None
plt.rcParams["figure.figsize"] = (10,6)
plt.rcParams['font.weight'] = 'bold'
#plt.rcParams['font.size'] = 22

weather.head()
```

```python
weather.info()
```

```python
weather.describe()
```

```python
weather['Date'] = pd.to_datetime(weather['Date'])
weather['Date'] 
```

```python
weather["year_month"] = weather['Date'].dt.strftime("%Y-%m")
weather["year"] = weather['Date'].dt.year
weather["month"] = weather['Date'].dt.month
```

```python
weather.head(1)
```

```python
# pop() method works for a single column and returns a Series

year_month = weather.pop("year_month")
year = weather.pop("year")
month = weather.pop("month")

# Let's insert the columns we have popp

weather.insert(1, "year_month", year_month)
weather.insert(2, "year", year)
weather.insert(3, "month", month)

weather.head()
```

```python
weather = weather[weather["year"]==2016]
```

```python
weather.head()
```

```python
weather.shape
```

```python
weather["Location"].unique()
```

```python
albury_weather = weather[weather["Location"]=="Albury"]
portland_weather = weather[weather["Location"]=="Portland"]
```

```python
albury_weather.head(1)
```

```python
portland_weather.head(1)
```

**Instructions**

- Create a multiple-line in the same graph with functional method.
- Plot MaxTemp from albury_weather against the month.
- Plot MaxTemp from portland_weather against the month.
- Set the x-axis label to 'Month'.
- Set the y-axis label to 'Maximum Temperature'
- Set title to 'Albury vs Portland Weather'- 
- Pass the ``color`` key-word arguments to these commands to set the color of the Albury data to blue ('b') and the Portland data to red ('r').
- Pass the ``marker`` key-word arguments to these commands to set the Albury data to circle markers ('o') and the Portland markers to triangles pointing downwards ('v').
- Pass the ``linestyle`` key-word argument to use dashed lines for the data from both cities ('--').

```python
# CODE HERE to get below output
plt.figure(figsize=(10,6))
plt.plot(albury_weather.groupby('month')['MaxTemp'].mean().index, albury_weather.groupby('month')['MaxTemp'].mean().values, "bo", ls="--", lw =3, markersize=10)
plt.plot(portland_weather.groupby('month')['MaxTemp'].mean().index, portland_weather.groupby('month')['MaxTemp'].mean().values, "rv", ls="--", lw=3, markersize=10)
plt.title('Albury vs Portland Weather', fontsize=20)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Maximum Temperature', fontsize=15)
```

**Instructions**

- Create a multiple-line in the same graph with object oriented method.
- Plot MinTemp from albury_weather against the month.
- Plot MinTemp from portland_weather against the month.
- Set the x-axis label to 'Month'.
- Set the y-axis label to 'Minimum Temperature'
- Set title to 'Albury vs Portland Weather'- 
- Pass the ``color`` key-word arguments to these commands to set the color of the Albury data to green ('g') and the Portland data to yellow ('y').
- Pass the ``marker`` key-word arguments to these commands to set the Albury data to x markers ('x') and the Portland markers to triangles pointing upwards ('^').
- Pass the ``linestyle`` key-word argument to use solid line for the data from both cities ('-').

```python
# CODE HERE to get below output
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(albury_weather.groupby('month')['MinTemp'].mean().index, albury_weather.groupby('month')['MinTemp'].mean().values, "gx", ls="--", lw =3, markersize=10)
ax.plot(portland_weather.groupby('month')['MinTemp'].mean().index, portland_weather.groupby('month')['MinTemp'].mean().values, "y^", ls="--", lw=3, markersize=10)
ax.set_title('Albury vs Portland Weather', fontsize=20)
ax.set_xlabel('Month', fontsize=15)
ax.set_ylabel('Minimum Temperature', fontsize=15);
```

**Instructions**

- Create a subplots with 2 rows and 2 columns with functional method.
- Addressing the top left as index (2, 2, 1) plot the Albury WindSpeed9am.
- Addressing the top right as index (2, 2, 2) plot the Portland WindSpeed9am.
- Addressing the bottom left as index (2, 2, 3) plot the Albury WindSpeed3pm.
- Addressing the bottom right as index (2, 2, 4) plot the Portland WindSpeed3pm.

```python
# CODE HERE to get below output
plt.subplot(2,2,1)
plt.title('Albury', fontsize = 20)
plt.ylabel('WindSpeed9am', fontsize = 15)
plt.plot(albury_weather.groupby('month')['WindSpeed9am'].mean(), lw =3)

plt.subplot(2,2,2)
plt.title('Portland', fontsize = 20)
plt.plot(portland_weather.groupby('month')['WindSpeed9am'].mean(), lw =3)

plt.subplot(2,2,3)
plt.ylabel('WindSpeed3pm', fontsize = 15)
plt.plot(albury_weather.groupby('month')['WindSpeed3pm'].mean(), lw =3)

plt.subplot(2,2,4)
plt.plot(portland_weather.groupby('month')['WindSpeed3pm'].mean(), lw =3)

plt.tight_layout()
```

**Instructions**

- Create a subplots with 2 rows and 2 columns with object oriented method.
- Addressing the top left as index (2, 2, 1) plot the Albury Humidity9am.
- Addressing the top right as index (2, 2, 2) plot the Portland Humidity9am.
- Addressing the bottom left as index (2, 2, 3) plot the Albury Humidity3pm.
- Addressing the bottom right as index (2, 2, 4) plot the Portland Humidity3pm.

```python
# CODE HERE to get below output
fig, ax = plt.subplots(nrows= 2, ncols = 2)

ax[0][0].set_title('Albury', fontsize = 20)
ax[0][0].set_ylabel('WindSpeed9am', fontsize = 15)
ax[0][0].plot(albury_weather.groupby('month')['WindSpeed9am'].mean(), lw =3)

ax[0][1].set_title('Portland', fontsize = 20)
ax[0][1].plot(portland_weather.groupby('month')['WindSpeed9am'].mean(), lw =3)

ax[1][0].set_ylabel('WindSpeed3pm', fontsize = 15)
ax[1][0].plot(albury_weather.groupby('month')['WindSpeed3pm'].mean(), lw =3)

ax[1][1].plot(portland_weather.groupby('month')['WindSpeed3pm'].mean(), lw =3)

fig.tight_layout()
```

```python
# CODE HERE to get below output
fig = plt.figure()

ax1 = fig.add_axes([0.0, 0.0, 0.6, 1]) # left, bottom, width, height (range 0 to 1)
ax1.set_xlabel("Month", fontsize=12)
ax1.set_ylabel("Minimum Temperature", fontsize=12)
ax1.plot(albury_weather.groupby('month')['MinTemp'].mean().index, albury_weather.groupby('month')['MinTemp'].mean().values, lw =3)

ax2 = fig.add_axes([0.9, 0.75, 0.2, 0.2])
ax2.set_ylabel("Cloud9am")
ax2.plot(albury_weather.groupby('month')['Cloud9am'].mean().index, albury_weather.groupby('month')['Cloud9am'].mean().values, lw =3)

ax3 = fig.add_axes([0.9, 0.50, 0.2, 0.2])
ax3.set_ylabel("Pressure9am")
ax3.plot(albury_weather.groupby('month')['Pressure9am'].mean().index, albury_weather.groupby('month')['Pressure9am'].mean().values, lw =3)

ax4 = fig.add_axes([0.9, 0.25, 0.2, 0.2])
ax4.set_ylabel("Humidity9am")
ax4.plot(albury_weather.groupby('month')['Humidity9am'].mean().index, albury_weather.groupby('month')['Humidity9am'].mean().values, lw =3)

ax5 = fig.add_axes([0.9, 0, 0.2, 0.2])
ax5.set_ylabel("WindSpeed9am")
ax5.set_xlabel("Month")
ax5.plot(albury_weather.groupby('month')['WindSpeed9am'].mean().index, albury_weather.groupby('month')['WindSpeed9am'].mean().values, lw =3)

plt.show()
```

# BONUS

**matplotlib.pyplot.twinx() Function**

The **twinx()** function in pyplot module of matplotlib library is used to make and return a second axes that shares the x-axis.

[Source](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/two_scales.html)

```python
albury_weather.groupby("month")["Pressure9am"].mean()
```

```python
albury_weather.groupby("month")["MinTemp"].mean()
```

```python
fig, ax = plt.subplots()

ax.plot(albury_weather["month"].unique(), albury_weather.groupby("month")["MinTemp"].mean())
ax.set_title("Albury")
ax.set_xlabel("Month")
ax.set_ylabel("MinTemp")

ax.plot(albury_weather["month"].unique(), albury_weather.groupby("month")["Pressure9am"].mean())
ax.set_title("Albury")
ax.set_xlabel("Month")
ax.set_ylabel("Pressure9am")

plt.tight_layout()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16,8))

ax[0].plot(albury_weather["month"].unique(), albury_weather.groupby("month")["MinTemp"].mean())
ax[0].set_title("Albury")
ax[0].set_xlabel("Month")
ax[0].set_ylabel("MinTemp")

ax[1].plot(albury_weather["month"].unique(), albury_weather.groupby("month")["Pressure9am"].mean())
ax[1].set_title("Albury")
ax[1].set_xlabel("Month")
ax[1].set_ylabel("Pressure9am")

plt.tight_layout()
```

```python
fig, ax = plt.subplots()

ax.plot(albury_weather["month"].unique(), albury_weather.groupby("month")["MinTemp"].mean())
ax.set_title("Albury")
ax.set_xlabel("Month")
ax.set_ylabel("MinTemp")

ax2 = ax.twinx() # The twinx() function is used to make and return a second axes that shares the x-axis.

ax2.plot(albury_weather["month"].unique(), albury_weather.groupby("month")["Pressure9am"].mean())
ax2.set_title("Albury")
ax2.set_xlabel("Month")
ax2.set_ylabel("Pressure9am")

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(albury_weather["month"].unique(), albury_weather.groupby("month")["MinTemp"].mean(), color="blue")
ax.set_title("Albury")
ax.set_xlabel("Month")
ax.set_ylabel("MinTemp", color="blue")

ax2 = ax.twinx()

ax2.plot(albury_weather["month"].unique(), albury_weather.groupby("month")["Pressure9am"].mean(), color="red")
ax2.set_title("Albury")
ax2.set_xlabel("Month")
ax2.set_ylabel("Pressure9am", color="red")

plt.show()
```

**Axes.tick_params(axis='both', **kwargs)**, change the appearance of ticks, tick labels, and gridlines.

[Source](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.tick_params.html)

```python
fig, ax = plt.subplots()

ax.plot(albury_weather["month"].unique(), albury_weather.groupby("month")["MinTemp"].mean(), color="blue")
ax.set_title("Albury")
ax.set_xlabel("Month")
ax.set_ylabel("MinTemp", color="blue")
ax.tick_params("y", colors="blue")

ax2 = ax.twinx()

ax2.plot(albury_weather["month"].unique(), albury_weather.groupby("month")["Pressure9am"].mean(), color="red")
ax2.set_title("Albury")
ax2.set_xlabel("Month")
ax2.set_ylabel("Pressure9am", color="red")
ax2.tick_params("y", colors="red")

plt.show()
```
